```python
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
img = image.load_img('/content/pexels-chevonrossouw-2558605.jpg',target_size=(200,200))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(img)
```

```python
type(img)
```

```python
from pathlib import WindowsPath
datagen = ImageDataGenerator(
          rotation_range=30,
          shear_range=0.2,
          zoom_range=0.2,
          horizontal_flip=True,
          width_shift_range=0.2,
          height_shift_range=0.2,
          fill_mode='nearest'
)
```

```python
img = image.img_to_array(img)
```

```python
type(img)
```

```python
img.shape
```

```python
input_batch=img.reshape(1,200,200,3)
```

```python
import os
```

```python
os.makedirs('aug', exist_ok=True)

i = 0
for batch in datagen.flow(input_batch,
                          batch_size=1,
                          save_to_dir='aug',
                          save_prefix='aug',
                          save_format='jpeg'):

    i += 1
    if i == 10:
        break
```
